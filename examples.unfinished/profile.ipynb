```python
import xarray as xr
import xroms
import matplotlib.pyplot as plt
```

```python
A = xroms.roms_dataset('ocean_avg_0014.nc')
i, j = 100, 110
tprofile0 = A.temp.isel(ocean_time=0, xi_rho=i, eta_rho=j)
sprofile0 = A.salt.isel(ocean_time=0, xi_rho=i, eta_rho=j)
depth = -A.z_rho.isel(xi_rho=i, eta_rho=j)
```

```python
coords={'depth': depth.values}
tprofile = xr.DataArray(tprofile0, dims=['depth'], coords=coords)
sprofile = xr.DataArray(sprofile0, dims=['depth'], coords=coords)
```

```python
fig, axes = plt.subplots(figsize=(6, 6), ncols=2, sharey=True)
tprofile.plot(y='depth', yincrease=False, ax=axes[0])

plt.subplot(1,2,2)
sprofile.plot(y='depth', yincrease=False);
plt.grid(True)
plt.title('Salinity');
```

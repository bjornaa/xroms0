```python
import matplotlib.pyplot as plt
import xroms
%matplotlib inline
```

```python
A = xroms.roms_dataset('ocean_avg_0014.nc')
```

```python
SST = A.temp.isel(time=0, s_rho=-1)
SST = SST.where(A.mask_rho > 0)
SST.plot()
plt.axis('image');
```

```python
# Temperature at 50 meter at first timeframe

T50 = xroms.zslice(A.temp.isel(time=0), 50)
```

```python
T50.plot.contourf()
M = A.mask_rho.where(A.mask_rho < 1)
M.plot(add_colorbar=False, cmap='PuRd_r')
plt.axis('image');
```

```python
# Check on zslice
V = xroms.zslice(A.z_rho, 50)
print(V.min(), V.mean(), V.max())
```

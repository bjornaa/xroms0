```python
import matplotlib
import matplotlib.pyplot as plt
import xarray as xr
%matplotlib inline
```

```python
#
# Prepare data
#
A = xr.open_dataset('ocean_avg_0014.nc')

H = A.h      # Get the bottom topography
H = H.where(A.mask_rho > 0)    # Mask out land
```

```python
# Make a quick and dirty plot
H.plot();
```

```python
# Improve the plot

# Choose levels to make shallower features visible.
levels = [10, 20, 50, 100, 200, 500, 1000, 2000]
# Deeper = darker colour.
cmap = matplotlib.cm.viridis_r

H.plot(levels=levels, cmap=cmap)
# Add isobaths
H.plot.contour(levels=levels, colors='black', linewidths=0.5)
plt.axis('image');  # Correct aspect ratio  
```

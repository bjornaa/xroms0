```python
# import numpy as np
import matplotlib.pyplot as plt
# import xarray as xr
import cmocean.cm as cmo
import xroms
```

```python
# Initiate the Dataset
A = xroms.roms_dataset("ocean_avg_0014.nc")
```

```python
T = A.salt.isel(ocean_time=0, xi_rho=100)
```

```python
salinity_levels=[30, 33, 34, 34.5, 34.8, 34.9, 35.0, 35.1, 35.2]
T.plot.pcolormesh('lat_rho', 'z_rho', levels=salinity_levels, cmap=cmo.haline_r)
plt.ylim(-200, 0)
```
